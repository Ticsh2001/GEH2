```python
def pow_expr(a, b):
    return f"POW({a},{b})"


def poly_expr(pi_expr, x_expr, terms):
    """
    Σ n * pi^I * x^J
    """
    parts = []
    for I, J, n in terms:
        term = f"{n}"
        if I != 0:
            term += f"*{pow_expr(pi_expr, I)}"
        if J != 0:
            term += f"*{pow_expr(x_expr, J)}"
        parts.append(term)
    return "(" + " + ".join(parts) + ")"


def generate_T_region2_expression():
    # symbols used in final expression
    p = "p"                     # MPa
    s = "s"                     # J/(kg*K)

    # --- subregion 2a ---
    sigma_2a = f"({s}/2000)"
    x_2a = f"({sigma_2a}-2)"
    T2a = poly_expr(p, x_2a, TERMS_2A)

    # --- subregion 2b ---
    sigma_2b = f"({s}/785.3)"
    x_2b = f"(10-{sigma_2b})"
    T2b = poly_expr(p, x_2b, TERMS_2B)

    # --- subregion 2c ---
    sigma_2c = f"({s}/2925.1)"
    x_2c = f"(2-{sigma_2c})"
    T2c = poly_expr(p, x_2c, TERMS_2C)

    # --- logical structure ---
    expr = (
        "WHEN("
        "(p<=4),"
        f"{T2a},"
        "WHEN("
        "((p>4) AND (s>=5850)),"
        f"{T2b},"
        f"{T2c}"
        ")"
        ")"
    )

    return expr


# ---- generate expression ----
expression = generate_T_region2_expression()
print(expression)
```
