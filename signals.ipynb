```python
import pandas as pd
import os
import json

def find_system_and_device(tag, systems_dict, devices_dict):
    tag_upper = str(tag).upper()
    
    # Ищем систему
    system_found = 'Unknown'
    system_code_found = None
    for system_code, system_name in systems_dict.items():
        if system_code in tag_upper:
            system_found = system_name
            system_code_found = system_code
            break
    
    # Если систему не нашли - выходим
    if system_found == 'Unknown':
        return 'Unknown', 'Unknown', 'Unknown', "Unknown"
    
    # Удаляем код системы из тега
    tag_without_system = tag_upper.replace(system_code_found, '', 1)
    
    # Ищем устройство в оставшейся части
    device_found = 'Unknown'
    device_code_found = 'Unknown'
    for device_code, device_name in devices_dict.items():
        if device_code in tag_without_system:
            device_code_found = device_code
            device_found = device_name
            break
    
    return system_found, device_found, system_code_found, device_code_found

with open('systems.json', 'r') as file:
    sys_data = json.load(file)

st_df = pd.read_csv(os.path.join('tec 20', 'tec20_analog_gt_st.csv'), sep=';')[['Tagname', 'TagId', 'Description','Engineering Unit']]
st_df[['system', 'device', 'sys', 'dev']] = st_df['Tagname'].apply(lambda x: pd.Series(find_system_and_device(x, sys_data['systems'], sys_data['devices'])))
st_df
```
